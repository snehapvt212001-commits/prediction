```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_cols = ['education', 'marital', 'month', 'poutcome', 'contact', 'job']
education_order = ['unknown', 'primary', 'secondary', 'tertiary']
month_order = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
poutcome_order = ['failure', 'nonexistent', 'success']
marital_order = ['divorced', 'married', 'single']
contact_order = ['cellular', 'telephone','unknown']
job_order = ['admin.', 'blue-collar', 'entrepreneur', 'housemaid', 'management', 'retired', 'self-employed', 'services', 'student', 'technician', 'unemployed', 'unknown']
category_orders = {
    'education': education_order,
    'marital': marital_order,
    'month': month_order,
    'poutcome': poutcome_order,
    'contact': contact_order,
    'job': job_order
}
oe = OrdinalEncoder(categories=[category_orders[col] for col in ordinal_cols],
                    handle_unknown='use_encoded_value',
                    unknown_value=-1,
                    dtype=int)
bank[ordinal_cols] = oe.fit_transform(bank[ordinal_cols])
print(bank.head())
```

```python
ip=bank.drop("y",axis=1)
op=bank.y
```

```python
ip.head()
```

```python
ip.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train.head()
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
pred=gnb.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
ac=accuracy_score(pred,y_test)
re=recall_score(pred,y_test)
pr=precision_score(pred,y_test)
f1=f1_score(pred,y_test)
print('accuracy:',ac)
print('recall;',re)
print('precision',pr)
print('f1_score',f1)
```

```python

df=pd.DataFrame({'y_test':list(y_test),'Prediction':pred})
df
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(pred,y_test)
cm
```
