```python
def llm_normalize_to_json(text):
    t = text.lower()

    name_match = re.search(r'‡¨Æ‡≠ã ‡¨®‡¨æ‡¨Æ (\S+)|my name is (\S+)', t)
    name = name_match.group(1) or name_match.group(2) if name_match else "User"

    age = word_to_number(t) or 25

    income_match = re.search(r'‡¨Ü‡≠ü ([\w\s]+)|income ([\w\s]+)', t)
    income = word_to_number(income_match.group(1)) if income_match else 150000

    occupation = "unemployed" if "‡¨¨‡≠á‡¨ï‡¨æ‡¨∞" in t else "employed"
    gender = "female" if "‡¨Æ‡¨π‡¨ø‡¨≥‡¨æ" in t else "male"
    residence = "rural" if "‡¨ó‡≠ç‡¨∞‡¨æ‡¨Æ" in t else "urban"

    return {
        "name": name.capitalize(),
        "age": age,
        "income": income,
        "occupation": occupation,
        "gender": gender,
        "residence": residence
    }
```

```python
# Browser-based audio recording for Colab
from IPython.display import Javascript, display
from google.colab import output
import base64
import numpy as np
from scipy.io.wavfile import write

RECORD_SECONDS = 30
SAMPLE_RATE = 16000

js = f"""
const sleep = time => new Promise(resolve => setTimeout(resolve, time));

async function record() {{
  const stream = await navigator.mediaDevices.getUserMedia({{ audio: true }});
  const mediaRecorder = new MediaRecorder(stream);
  let chunks = [];

  mediaRecorder.ondataavailable = e => chunks.push(e.data);
  mediaRecorder.start();

  await sleep({RECORD_SECONDS * 1000});
  mediaRecorder.stop();

  await new Promise(resolve => mediaRecorder.onstop = resolve);

  const blob = new Blob(chunks);
  const reader = new FileReader();
  reader.readAsDataURL(blob);

  reader.onloadend = () => {{
    const base64data = reader.result.split(',')[1];
    google.colab.kernel.invokeFunction(
      'notebook.save_audio',
      [base64data],
      {{}}
    );
  }};
}}

record();
"""
display(Javascript(js))

def save_audio(base64data):
    audio = base64.b64decode(base64data)
    with open("user.wav", "wb") as f:
        f.write(audio)

output.register_callback('notebook.save_audio', save_audio)
```
