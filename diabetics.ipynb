```python
# input and output creation
ip=dia.drop("Outcome",axis=1)
op=dia.Outcome
```

```python
ip.head()
```

```python
op.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.25)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc=accuracy_score(pred,y_test)
re=recall_score(pred,y_test)
pr=precision_score(pred,y_test)
f1=f1_score(pred,y_test)
print("accuracy score",acc)
print("recall score",re)
print("precision score",pr)
print("f1 score",f1)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(pred,y_test)
cm
```

```python
cmd=ConfusionMatrixDisplay(cm)
cmd.plot(cmap="Blues")
plt.show()
```
